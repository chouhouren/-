```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import KFold
from sklearn.utils import check_array
```

```python
df = pd.read_excel("/Users/chouhouren/Desktop/論文資料/建模型資料/df_直線模型.xlsx")
```

```python
x = np.array(df.iloc[:,2:])
y = np.array(df.iloc[:,1])
x = np.array([np.concatenate((v,[1])) for v in x])
```

```python
linreg = LinearRegression()
```

```python
# Now let's compute RMSE using 10-fold x-validation
kf = KFold(len(x), n_folds=10)
xval_err = 0
for train,test in kf:
    linreg.fit(x[train],y[train])
    # p = np.array([linreg.predict(xi) for xi in x[test]])
    p = linreg.predict(x[test])
    e = p-y[test]
```

```python
y_test=pd.DataFrame(y[test])
y_test=np.exp(y_test)
y_test
```

```python
p = pd.DataFrame(p)
p = np.exp(p)
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
MAPE=mean_absolute_percentage_error(y_test,p).round(2)
MAPE
```

```python
y_nagative_10 = np.array(y_test - y_test*0.05)
y_positive_10 = np.array(y_test + y_test*0.05)
y_predict     = np.array(p)

y_n10=pd.DataFrame(y_nagative_10,columns=['y-10%'])

y_predict=pd.DataFrame(y_predict,columns=['y'])

y_p10=pd.DataFrame(y_positive_10,columns=['y+10%'])



hit_rate = pd.concat([y_n10,y_predict,y_p10],axis=1)


命中值 =[(hit_rate['y']>hit_rate['y-10%'])&(hit_rate['y']<hit_rate['y+10%'])]


choices = [1]
hit_rate['命中值'] = np.select(命中值,choices,default=0)
HIT_rate = (hit_rate['命中值'].sum()/len(y_test)*100).round(2)
HIT_rate.round(2)
```
