```python
def distance_caculation(x1,y1,x2,y2):
    R = 6373000

    lat1 = radians(x1)
    lng1 = radians(y1)
    
    lat2 = radians(x2)
    lng2 = radians(y2)

    dlng = lng2 - lng1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlng / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c

    return distance      
```

```python
#1.捷運出口
x_1 = df_實價登錄座標["lat"]
y_1 = df_實價登錄座標["lng"]
x_2 = df_迎毗設施_捷運站出口["lat"]
y_2 = df_迎毗設施_捷運站出口["lng"]


#call the function

dflist_distance_2 = []
for a1,n1 in zip(x_1,y_1):
    for a2,n2 in zip(x_2,y_2):
        
        x1 =a1
        y1 =n1
        x2 =a2
        y2 =n2
        result = distance_caculation(x1,y1,x2,y2)
        
        dflist_distance_2.append(result)
        
df_捷運站出口距離 = pd.DataFrame(dflist_distance_2)



df_捷運站出口距離_重組 = df_捷運站出口距離.values.reshape((13626,363))

df_捷運站出口距離_重組_pd = pd.DataFrame(df_捷運站出口距離_重組.min(axis=1),columns=["捷運出口"])

df_捷運站出口距離_重組_pd.head()
```
