```python
from math import sin, cos, sqrt, atan2, radians
import pandas as pd
import numpy as np

df_實價登錄座標         = pd.read_excel("/Users/chouhouren/Desktop/實價登錄清理.xlsx")

df_鄰避設施_加油站     = pd.read_excel("/Users/chouhouren/Desktop/論文資料/鄰避設施/鄰避設施.xlsx",sheet_name="加油站")

df_鄰避設施_電塔        = pd.read_excel("/Users/chouhouren/Desktop/論文資料/鄰避設施/鄰避設施.xlsx",sheet_name="電塔")

df_鄰避設施_污水處理廠   = pd.read_excel("/Users/chouhouren/Desktop/論文資料/鄰避設施/鄰避設施.xlsx",sheet_name="污水處理廠")

df_鄰避設施_垃圾焚化廠   = pd.read_excel("/Users/chouhouren/Desktop/論文資料/鄰避設施/鄰避設施.xlsx",sheet_name="垃圾焚化廠")

df_鄰避設施_飛機場      = pd.read_excel("/Users/chouhouren/Desktop/論文資料/鄰避設施/鄰避設施.xlsx",sheet_name="飛機場")

df_鄰避設施_變電所      = pd.read_excel("/Users/chouhouren/Desktop/論文資料/鄰避設施/鄰避設施.xlsx",sheet_name="變電所")

df_鄰避設施_殯儀館      = pd.read_excel("/Users/chouhouren/Desktop/論文資料/鄰避設施/鄰避設施.xlsx",sheet_name="殯儀館")

df_鄰避設施_醫院      = pd.read_excel("/Users/chouhouren/Desktop/論文資料/鄰避設施/鄰避設施.xlsx",sheet_name="醫院")


```

```python
#1.加油站
x_1 = df_實價登錄座標["lat"]
y_1 = df_實價登錄座標["lng"]
x_2 = df_鄰避設施_加油站["加油站lat"]
y_2 = df_鄰避設施_加油站["加油站lng"]

def distance_caculation(x1,y1,x2,y2):
    R = 6373000

    lat1 = radians(x1)
    lng1 = radians(y1)
    
    lat2 = radians(x2)
    lng2 = radians(y2)

    dlng = lng2 - lng1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlng / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c

    return distance 
    
    
#call the function

dflist_distance_2 = []
for a1,n1 in zip(x_1,y_1):
    for a2,n2 in zip(x_2,y_2):
        
        x1 =a1
        y1 =n1
        x2 =a2
        y2 =n2
        result = distance_caculation(x1,y1,x2,y2)
        
        dflist_distance_2.append(result)
        
        
df_加油站距離 = pd.DataFrame(dflist_distance_2)


df_加油站距離_重組 = df_加油站距離.values.reshape((13626, 75))


df_加油站距離_重組_pd = pd.DataFrame(df_加油站距離_重組.min(axis=1),columns=["加油站"])

df_加油站距離_重組_pd.head()        
```
